# CsPbBr$_3$

## Slab thickness analysis and van der Waals corrections

The van der Waals strength between $CsPbBr_3$ layers was proved as a non-negligible interaction. So, in order to comprehend the influence of van der Waals interactions in Ab Initio surface energy calculations, we carried out two differents van de Waals corrections, called by grimme-d2 and grimme-d3. Also, we repeat the same methodology without any van der Waals correction, as a control sample.
